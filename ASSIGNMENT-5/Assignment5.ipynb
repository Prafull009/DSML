```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
admit = pd.read_csv("/content/admission - admission.csv")
admit.head
```

```python
# Scatter plot
plt.scatter(admit["CGPA"], admit["Chance of Admit"])
plt.xlabel("CGPA")
plt.ylabel("Chance of Admit")
plt.title("CGPA vs Chance of Admit")
plt.show()
```

```python
# Bar Graph
plt.bar(admit["University Rating"], admit["Serial No."])
plt.xlabel("University Rating")
plt.ylabel("No. of Admission")
plt.title("University Rating vs No. of Admission")
plt.show()
```

```python
# Boxplot
sns.boxplot(x = 'Research', y = 'Chance of Admit', data = admit)
```

```python
# Histogram
plt.hist(admit["LOR"], histtype = "bar")
plt.xlabel("LOR")
plt.ylabel("No. of Applicants")
plt.show()
```

```python
# Line Graph
plt.plot(admit["Chance of Admit"], admit["Research"])
plt.show()
```

```python
# Pie Chart
l = [1, 0]
plt.pie(admit[""], labels = admit["University Rating"])
plt.show()
```

```python
sns.kdeplot(data=admit, x="University Rating")
```

```python
plt.imshow(admit["CGPA"], interpolation = 'nearest', cmap = 'Blues')
plt.title("Heat Map")
plt.colorbar()
plt.show()
```

```python
sns.countplot(x = "Research", data = admit)
```
