```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
points =[(0.1, 0.6), (0.15, 0.71), (0.08, 0.9), (0.16, 0.85), (0.2, 0.3), (0.25, 0.5), (0.24, 0.1), (0.3, 0.2)]
centroids = [(0.1, 0.6), (0.3, 0.2)]
```

```python
X = [x[0] for x in points]
Y = [x[1] for x in points]
fig = plt.figure(0)
plt.title('Scatter Plot of points')
plt.scatter(X, Y)
plt.show()
```

```python
classifier = KMeans(n_clusters = 2, init = centroids, n_init = 1).fit(points)

centroids = classifier.cluster_centers_
labels = classifier.labels_
```

```python
for i, centroid in enumerate(centroids):
  pointsInClusters = [points[j] for j in range(len(points)) if labels[j] == i]
  print(f"Cluster {i+1}: Centroid {centroid}- Points {pointsInClusters}")
```

```python
n_clusters = 2
cluster_data = [[] for _ in range(n_clusters)]
for i, point in enumerate(points):
    cluster_data[labels[i]].append(point)

colors = ['red', 'green']
markers = ['o', 's']

for i in range(n_clusters):
    x, y = zip(*cluster_data[i])
    plt.scatter(x, y, label=f'Cluster {i+1}', c=colors[i], marker=markers[i])
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=100, linewidths=2, c='black', label='Centroids')

plt.title('K-Means Clustering')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
P6 = [0.25,0.5]
c = classifier.predict([P6])[0]
print("P6 belongs to ", (c+1))
```
